# Interpolation
## 1. Interpolating with exponentials

Consider a function of the following form:
$$
e_{m}(x) = \sum_{i=0}^{m} a_{i}e^{ix}
$$

We want to find some integer m satisfying
$$
e_{m}(x_{i}) = y_{i} \forall i = 0...n
$$

Or in other words for any set of n+1 points with distinct $x_{i}$, we want to find an equation that produces $y_{i}$.

To start, set up a system of equations for each x, y pair:

$$
\begin{bmatrix}
1& e^{x_{0}}& e^{2x_{0}}& ...& e^{nx_{0}} \\
1& e^{x_{1}}& e^{2x_{1}}& ...& e^{nx_{1}}\\
1& e^{x_{2}}& e^{2x_{2}}& ...& e^{nx_{2}}\\
...& ...& ...& ...& ...& \\
1& e^{x_{n}}& e^{2x_{n}}& ...& e^{nx_{n}}
\end{bmatrix}
*
\begin{bmatrix}
a_{0} \\
a_{1} \\
a_{2} \\
... \\
a_{n}
\end{bmatrix}
=
\begin{bmatrix}
y_{0} \\
y_{1} \\
y_{2} \\
...   \\
y_{n}
\end{bmatrix}
$$

This matrix is in the form of the Vandermode Matrix, which has a non-zero determinate if each $\alpha_{i}$ is unique for each row (see https://en.wikipedia.org/wiki/Vandermonde_matrix). We can see that each row is of the form $[1, (e^{x_i})^{1}, (e^{x_i})^{2} ...]$ and since each $x_{i}$ is unique by construction, the conditions for a non-zero determinate are met. Therefore the matrix is invertible and the system has a solution for any set of x, y pairs.

## 2. Convergence of interpolating polynomial

Let $$\psi_n(x) = (x-x_0)(x-x_1)...(x-x_{n-1})(x-b)$$
<br\>
$$= (x)(x-h)(x-2h)...(x-nh)$$

$$ \left|\psi_n(x)\right| \leq n!h^{n+1}  \,\,\,\,\,  \forall x \in [0,a]$$
<br\>
<br\>
In general the error E in polynomial interpolation is given by
$$E(x) = \frac{\psi_n(x)}{(n+1)!}*f^{n+1}(\zeta)$$
<br\>
this implies
$$\underset{x\in [0,a]}{\max} \left|e^{x} - p_n(x)\right| \leq \frac{c_{n+1}}{(n+1)!}* \underset{x\in [0,a]}{\max} \left|\psi_n(x)\right|$$
<br\>
with
<br\>
$$ c_{n+1} = \underset{x\in [0,a]}{\max} \left|f^{n+1}(x)\right|$$
<br\>
Since $e^x$ equals its derivatives $c_{n+1} = \underset{x\in [0,a]}{\max} \left|e^x\right|$ and since a is just a finite real number $c_{n+1}$ is also just equal to some real number, call this k
<br\>
<br\>
<br\>
Now we have that $\left|\psi_n(x)\right| \leq n!h^{n+1}  \,\,\,\,\,  \forall x \in [0,a]$ and $c_{n+1}=k$ so,
<br\>
$$\underset{x\in [0,a]}{\max} \left|e^{x} - p_n(x)\right| \leq \frac{k}{(n+1)!}*n! h^{n+1} = \frac{k}{(n+1)}*h^{n+1}$$
Now since a is a number, as $n \rightarrow \infty$ h must become less than 1 but is always greater than zero, lets say for this sufficently large n, h=g so $g \in (0,1)$
<br\>
now
<br\>
$$\underset{x\in [0,a]}{\max} \left|e^{x} - p_n(x)\right| \leq \frac{k}{(n+1)}*g^{n+1}$$

now as since $g \in (0,1)$ as $n \rightarrow \infty$
<br\>
$$ \frac{k}{(n+1)}*g^{n+1}\rightarrow 0$$

## 3. Non-Linear Interpolation

Consider an interpolating function of the form:
$$p(x) = \frac{a + bx}{1 + cx}$$

Such a function must interpolate the 3 (x, y) pairs, with each x being distinct.

We want to show that a there exists a function of the form p(x) for all possible data points. If there does not exist such a function, then show the conditions that will ensure that a solution exists and that the solution is unique.

We can rewrite p(x):

$$y = \frac{a + bx}{1 + cx}$$
$$y = a + bx - cxy$$

We can now represent the three points as a system of equations:

$$
\begin{bmatrix}
1& x_{1}& x_{1}y_{1} \\
1& x_{2}& x_{2}y_{2} \\
1& x_{3}& x_{3}y_{3}
\end{bmatrix}
*
\begin{bmatrix}
a \\
b \\
c
\end{bmatrix}
=
\begin{bmatrix}
y_{1} \\
y_{2} \\
y_{3}
\end{bmatrix}
$$

Using row reduction, we can try to find the conditions for a solution:

$$
Begin:
\begin{bmatrix}
1& x_{1}& x_{1}y_{1}& y_{1} \\
1& x_{2}& x_{2}y_{2}& y_{2} \\
1& x_{3}& x_{3}y_{3}& y_{3}
\end{bmatrix}
$$
$$
R_{1} - R_{2} = R_{2}:
\begin{bmatrix}
1& x_{1}& x_{1}y_{1}& y_{1} \\
0& x_{1} - x_{2}& x_{1}y_{1} - x_{2}y_{2}& y_{1} - y_{2} \\
1& x_{3}& x_{3}y_{3}& y_{3}
\end{bmatrix}
$$
$$
R_{1} - R_{3} = R_{3}:
\begin{bmatrix}
1& x_{1}& x_{1}y_{1}& y_{1} \\
0& x_{1} - x_{2}& x_{1}y_{1} - x_{2}y_{2}& y_{1} - y_{2} \\
0& x_{1} - x_{3}& x_{1}y_{1} - x_{3}y_{3}& y_{1} - y_{3} \\
\end{bmatrix}
$$
$$
\frac{R_{2}}{x_{1} - x_{2}} = R_{2}:
\begin{bmatrix}
1& x_{1}& x_{1}y_{1}& y_{1} \\
0& 1& \frac{x_{1}y_{1} - x_{2}y_{2}}{x_{1} - x_{2}}& \frac{y_{1} - y_{2}}{x_{1} - x_{2}} \\
0& x_{1} - x_{3}& x_{1}y_{1} - x_{3}y_{3}& y_{1} - y_{3} \\
\end{bmatrix}
$$
$$
R_{2}({x_{1} - x_{3}}) - R_{3} = R_{3}:
\begin{bmatrix}
1& x_{1}& x_{1}y_{1}& y_{1} \\
0& 1& \frac{x_{1}y_{1} - x_{2}y_{2}}{x_{1} - x_{2}}& \frac{y_{1} - y_{2}}{x_{1} - x_{2}} \\
0& 0& (x_{1} - x_{3})\frac{x_{1}y_{1} - x_{2}y_{2}}{x_{1} - x_{2}} - (x_{1}y_{1} - x_{3}y_{3})& (x_{1} - x_{3})\frac{y_{1} - y_{2}}{x_{1} - x_{2}} - (y_{1} - y_{3}) \\
\end{bmatrix}
$$

At this point we have a system that is in reduced row echelon form, and the system only has a solution if the terms in row 3, column 3 and row 3, column 4 are equal.

So any set of x, y points that satisfy the following equation can be interpolated by a function a form of p(x):

$$
(x_{1} - x_{3})\frac{x_{1}y_{1} - x_{2}y_{2}}{x_{1} - x_{2}} - (x_{1}y_{1} - x_{3}y_{3}) = (x_{1} - x_{3})\frac{y_{1} - y_{2}}{x_{1} - x_{2}} - (y_{1} - y_{3})
$$

Let's try the following set of pairs {(1, $y_{1}$), (0, $y_{2}$), (-1, $y_{3}$)}:

We get
$$y_{1} - y_{3} = y_{1} - 2y_{2} + y{3}$$
$$y_{3} = y_{2}$$

Clearly there exist a $y_{3}$ and $y_{2}$ that are not equal, and therefore there does not exist a p(x) that can interpolate any set of x, y pairs.

Any 3 x,y pairs that satisfy the above equation with non-zero result is also unique because there are no free variables in the above system of equations.

## 2. Approximating cos(x)

$$

We will find the least squares approximation for some polynomial $a_0 + a_{1}x + a{2}x^2$.

$$
\left\lVert cos(x) - (a_0 + a_{1}x + a_{2}x^2) \right\rVert_{2}

## 1. Coding Simpson's Rule


```python
def simpson(f, a, b, n):
    if n % 2 == 1:
        raise Exception("invalid argument for n. must be an even number")
    
    accumulator = f(a)
    accumulator += f(b)
    
    delta = (b - a) / n
    
    for i in range(1, n):
        yi = f((i * delta) + a)
        accumulator += 2 * yi if (i % 2) == 0 else 4 * yi
    
    return accumulator * delta / 3

simpson(lambda x: 1/x, 1, 2, 100)
```

### (a)
The estimate of the definite integral using the midpoint rule is given by

$$ I_0=f(\frac{a+b}{2})(b-a)$$
<br/>
The error of approximating a definite integral using polynomial
interpolation over the set of points {x_0,x_1,...,x_n} is given by
<br/>
$$ E_0(f)=\int_{a}^{b}[f(x)-p(x)]dx = \int_{a}^{b}\Bigg[\frac{f^{n+1}(\eta)}{(n+1)!}\prod_{i=0}^{n}(x-x_i)\Bigg]dx $$
<br/>
Since we approximating the integral with a constant function, adding any point in $[a,b]$ and interpolating through it does not affect the the area. In order to get an expression for the error we need to interpolate through at least two points, so lets just double the midpoint so n=2 and
<br/>
$$ \prod_{i=0}^{2}(x-x_i) = (x-\frac{a+b}{2})(x-\frac{a+b}{2}) $$
<br/>
Now the error is given by
<br/>
$$ E_0(f)=\frac{{f}''(\eta)}{2}\int_{a}^{b}\Bigg(x-\frac{a+b}{2}\Bigg)^{2}dx $$
<br/>
$$ = \frac{{f}''(\eta)}{2} \frac{\big(x-\frac{a+b}{2}\big)^{3}}{3} \Big \vert_{a}^{b} $$
<br/>
$$ = \frac{{f}''(\eta)}{2} \frac{2(b-a)^{3}}{24} $$
<br/>
$$ = \frac{{f}''(\eta)}{24} (b-a)^{3} $$
<br/>

x_i = cos(\frac{\pi(2k - 1)}{2n}) \quad k = 1,..., n
$$

The function from (a) has a root(s) on interval \[-1, 1\] when:

$$
0 = cos(n*cos^{-1}(x))
$$

The $cos$ function is zero at equal to $\frac{pi}{2} * (2k + 1)$, where k is a non-zero integer
$$
\frac{\pi}{2} * (2k -1) = n * cos^{-1}(x)
$$
$$
\frac{\pi}{2n} * (2k -1) = cos^{-1}(x)
$$
$$
cos(\frac{\pi}{2n} * (2k -1)) = x
$$

Now we will only get distinct values when $0 < 2k - 1 < 2n$, which (since k is an integer) is equivilant to saying that k must be less than or equal to n.
